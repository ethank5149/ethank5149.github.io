##### Hard-Coding Labels

````python
class CategoryLabel:
    def __init__(self, label, score=0, color=None, color_pallete=default_colors):
        self.label = label
        self.score = score
        self._color = color
        self.color_pallete = color_pallete
        self.__repr__, self.__str__ = self.label, self.label
    
    @property
    def color(self):
        if self._color is None:
            self._color = self.color_pallete[self.score]
        return self._color
````
