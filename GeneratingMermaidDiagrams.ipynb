```sage
import base64
from PIL import Image
import requests, io
from IPython.display import SVG


def mermaid(graph_string, output_file='./out.png', background_color='darkred'):
    base64_string = base64.b64encode(graph_string.encode('ascii')).decode('ascii')
    inkd_graph = io.BytesIO(requests.get('https://mermaid.ink/img/' + base64_string + '?bgColor=!' + background_color).content)
    saved_graph = Image.open(inkd_graph)
    saved_graph.save(output_file)
    return saved_graph

def mermaid_svg(graph_string, output_file='./out.svg', background_color='darkred'):
    base64_string = base64.b64encode(graph_string.encode('ascii')).decode('ascii')
    inkd_graph = requests.get('https://mermaid.ink/svg/' + base64_string + '?bgColor=!' + background_color).content
    with open(output_file, 'wb') as f:
        f.write(inkd_graph)
    saved_graph = SVG(data=inkd_graph)
    return saved_graph
```
